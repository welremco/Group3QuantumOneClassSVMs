## Experiment 1: Repetitions of Circuit

```python
reps = [1, 2, 3]
results_reps = [0.7314758732081567, 0.7789473684210526, 0.668421052631579, 0.35096153846153844, 0.4708994708994709, 0.5410989010989011, 0.41781622632686466, 0.5238095238095238, 0.4825268817204302, 0.5438932372318261, 0.49567809239940386, 0.6117900790797988]

ds1_f1 = results_reps[:3]
ds2_f1 = results_reps[3:6]
ds3_f1 = results_reps[6:9]
ds4_f1 = results_reps[9:]

bar_width = 0.2

coeffs = [-1.5, -0.5, 0.5, 1.5]

def grouped_bars(ds):
    return [rep + bar_width*coeffs[ds] for rep in reps]

for ds, f1 in enumerate([ds1_f1, ds2_f1, ds3_f1, ds4_f1]):
    plt.bar(grouped_bars(ds), f1, width=bar_width, label=f'Dataset {ds+1}')
    
plt.xticks(reps)
plt.xlabel('Circuit repetitions')
plt.ylabel('Macro F1 score')
plt.title('Effect of Circuit Repetitions on Macro F1 Score')
plt.legend()
#plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left", borderaxespad=0.0)
plt.ylim(0.3)
plt.plot() 
```

## Experiment 4: Comparison to Classical SVMs

Compare best of classical methods to best of quantum methods

```python
kernel_list = ['rbf', 'linear', 'poly', 'sigmoid']
results_classical = [0.6749444780940843, 0.9171052631578946, 0.9171052631578946, 0.08695652173913042, 0.5330934649820098, 0.5881006864988558, 0.5881006864988558, 0.5881006864988558, 0.44148936170212766, 0.466172552142354, 0.4763827919227393, 0.4473684210526316, 0.5987654320987654, 0.537037037037037, 0.5330934649820098, 0.5330934649820098]

ds1_f1 = results_classical[:4]
ds2_f1 = results_classical[4:8]
ds3_f1 = results_classical[8:12]
ds4_f1 = results_classical[12:]

bar_width = 0.2

coeffs = [-1.5, -0.5, 0.5, 1.5]

def grouped_bars(ds):
    return [rep + bar_width*coeffs[ds] for rep in [1,2,3,4]]

for ds, f1 in enumerate([ds1_f1, ds2_f1, ds3_f1, ds4_f1]):
    plt.bar(grouped_bars(ds), f1, width=bar_width, label=f'Dataset {ds+1}')

plt.xticks([1,2,3,4], kernel_list)
plt.xlabel('Kernel')
plt.ylabel('Macro F1 score')
plt.title('Macro F1 Score for different classical Kernels')
plt.ylim(0.35)
plt.legend()
plt.show()
```

Reverse plot with datasets as x-axis

```python
kernels_f1 = []
for i in range(4):
    kernels_f1.append(results_classical[i::4])

bar_width = 0.2

coeffs = [-1.5, -0.5, 0.5, 1.5]

def grouped_bars(ds):
    return [rep + bar_width*coeffs[ds] for rep in [1,2,3,4]]

for ds, f1 in enumerate(kernels_f1):
    plt.bar(grouped_bars(ds), f1, width=bar_width, label=kernel_list[ds])

plt.xticks([1,2,3,4])
plt.xlabel('Dataset')
plt.ylabel('Macro F1 score')
plt.title('Macro F1 Score performance of different classical Kernels \n on each dataset')
plt.ylim(0.35)
plt.legend()
plt.show()
```

Find best for each dataset

```python
for i, results in enumerate([ds1_f1, ds2_f1, ds3_f1, ds4_f1], start=1):
    best = np.argmax(results)
    print(f'Dataset {i} best kernel: {kernel_list[best]} \t({results[best]*100:.2f}%)')
```

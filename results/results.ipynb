```python
import matplotlib.pyplot as plt
```

## Experiment 1: Repetitions of Circuit

```python
reps = [1, 2, 3]
ds1_precision_inliers = [0.9476, 0.9578, 0.9368]
ds1_precision_outliers = [0.5263, 0.6, 0.4]
ds1_recall_inliers = [0.7263, 0.7789, 0.6684]

ds1_f1_inliers = [0.7314, 0.7789, 0.6684]
ds2_f1_inliers = [0.3509, 0.4708, 0.5061]
ds3_f1_inliers = [0.4178, 0.5238, 0.4825]
ds4_f1_inliers = [0.5438, 0.4956, 0.6117]


#plt.plot(reps, ds1_precision_inliers, label='precision inliers')
#plt.plot(reps, ds1_precision_outliers, label='precision outliers')
#plt.plot(reps, ds1_recall_inliers, label='recall inliers')

#plt.plot(reps, ds1_f1_inliers, label='D1')
#plt.plot(reps, ds2_f1_inliers, label='D2')
#plt.plot(reps, ds3_f1_inliers, label='D3')
#plt.plot(reps, ds4_f1_inliers, label='D4')

bar_width = 0.2

coeffs = [-1.5, -0.5, 0.5, 1.5]

def grouped_bars(ds):
    return [rep + bar_width*coeffs[ds] for rep in reps]

for ds, f1 in enumerate([ds1_f1_inliers, ds2_f1_inliers, ds3_f1_inliers, ds4_f1_inliers]):
    plt.bar(grouped_bars(ds), f1, width=bar_width, label=f'Dataset {ds+1}')
    
plt.xticks(reps)
plt.xlabel('Circuit repetitions')
plt.ylabel('F1 score (inliers = positive class)')
plt.title('Effect of Circuit Repetitions on F1 Score')
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left", borderaxespad=0.0)
plt.ylim(0.25)
plt.plot() 
```

## Experiment 4: Comparison to Classical SVMs

Compare best of classical methods to best of quantum methods

```python
best_f1_classical = [0.9171, 0.5881, 0.4473, 0.5987]
best_f1_quantum = [0.7789, 0.5061, 0.5238, 0.6117]

datasets=[1,2,3,4]
bar_width = 0.2
coeffs = [-0.5, 0.5]
label = ['Classical', 'Quantum']

def grouped_bars(ds):
    return [dataset + bar_width*coeffs[ds] for dataset in datasets]

for ds, f1 in enumerate([best_f1_classical, best_f1_quantum]):
    plt.bar(grouped_bars(ds), f1, width=bar_width, label=label[ds])
    
plt.xticks([1,2,3,4], ['D1', 'D2', 'D3', 'D4'])
plt.xlabel('Datasets')
plt.ylabel('F1 score (inliers = positive class)')
plt.title('Comparison of F1 Score \n for Best Performing Classical and Quantum SVMs')
plt.legend()
plt.ylim(0.35)
plt.plot() 
```
